## Budapest interaktív kerületi szintű térkép
**Rövid leírás**: A cél egy kerületi szintű interaktív térkép elkészítése, ami a KSH-tól lekérdezett adatokat ábrázolja. A felhasználó külön évekhez és mutatókhoz kapcsolódó statisztikákat kérdezhet le. 

**adat források**
- [KSH éves Budapesti adatok kerületi bontásban](http://statinfo.ksh.hu/Statinfo/haDetails.jsp)

**begyűjtött adat struktúrája**
- *Táblák száma*: kb 20 külön tábla, amit letöltés után összemergelünk (sajnos a KSH nem jól optimalizál a bulk download-ra)
- *Adatok mennyisége*: 5-6 év * 20 (inkább kb. 17 releváns) adattábla * 23 kerület = 2760 adatpont
- *Lekérdezés menete*: Valószínűleg Keresési eredmények Beautiful Soup-ot kell majd használni az adatok letöltéséhez, mivel a KSH STADAT felületének nincsen API-ja (vagy még nem találtam).
 
**frissítési rendszeresség és a frissítés menete**
- Sajnos ezek az adatok (és bármi adat, amit már nem csináltak meg király térképes formában) csak évente frissülnek, így a lekérdezés is évente történne meg. Az első lekérdezéstől kezdve automatizálni szeretnénk a dolgot.
    - Adatokat lescrapeljük
    - Táblákaz mergeljük és updateljük 
    - Térképet updateljük új éves adatokkal
    
**milyen értékes információt lehet a begyűjtött anyagból egy dashboardon kommunikálni**
- User interakció az lenne, hogy legördülő menűben ki lehet választani a mutatót és az évet, így frissül a térkép.
- Extra külső tudást mi nem gondoltunk még, de ha van megfelelő adat új mutatók kialakítására ahhoz lehet statozunk.

## Rajkos ötlet - kurzusmap pro
**Rövid leírás**: Anna kurzustérképét fejlesztenénk tovább új feature-ökkel. Pl. top 5 kurzustárs, állandótanárok jelölése, vizuális jellegű felpimpelés, vagy akár valami prediktív modell használata (milyen kurzust érdemes felvenned, min kaphatsz kivállót stb.).

**adat források**
- Kurzusadatbázis, SZMT-s Fókák, Drive linkek

**begyűjtött adat struktúrája**
- Több Adattábla, viszont elég könnyen begyűjthetők a google sheetekről
- Kis adatbázisok, a kurzusadatbázis 300kb megfigyelés
- Google-Spredekből történő lekérdezésekre kell számítani 

**frissítési rendszeresség és a frissítés menete**
- Fél évente frissül a kurzusadatbázis, így nem kell gyors lekérdezésekre számítani

**milyen értékes információt lehet a begyűjtött anyagból egy dashboardon kommunikálni**
- Azt már tudjuk, hogy témák szerint ki milyen kurzusokra járt. Célunk itt ennek a kontentnek a továbbfejlesztése lenne, extra featureök hozzáadávál. 
- Mik lehetnek az extra featureök:
    - Kurzusmap vizuális felpimpelése

    - Állandó tanárok bejelülése
    - Minden kurzushoz hozzápakolni a tematikát, beszámolókat, drive mappát, stb.

## Matching platform különböző társadalmi vállalkozások összepakolására
**Rövid leírás**: Ez az ötlet a Covidból származik, viszont az alkalmazás általánosan is használható. Maradjunk a Covid példájánál. Nagyon sok egyetemistának/ középiskolásnak jó lenne, ha valaki tudna segíteni különböző tárgyak megtanulásában. Még ha erre van is alkalmas segítő, akkor is kell egy összekötő bürokratikus cucc, aki párosít. Az az ötletünk, hogy erre hoznánk létre egy felületet. Kicsit úgy képzeljük el mint a Tindert, csak itt tanárok és diákok találnak párokat egymásnak.

**Koncepció**: : Kitöltesz egy formot. Ezeket az adatokat eltárolja egy spread, ez a spread van meghívva a kódunkban. A párosítás után a pár mindkét tagja kap egy emailt, hogy its a match.

**begyűjtött adat struktúrája**
- Google spread
- Attól függ, hogy mennyien használják és milyen sok információt kérünk be egy embertől

**frissítési rendszeresség és a frissítés menete**
- Gyakran kellene frissíteni a rendszernek, mivel nem tudjuk, mikor érkezik be új input. 
- Az új adat egy új megfigyelést fog jelenteni, viszont a beérkező input potenciálisan több, már meglévő inputtal is matchel, úgyhogy érdemes a matchelésnél figyelembe venni, hogy ki kapott már korábbi párt.

**milyen értékes információt lehet a begyűjtött anyagból egy dashboardon kommunikálni**
- Kell benne user interakció, hiszen ki kell tölteni  kérdőívet, és ha megtörtént a találkozó, akkor még ott is lehet beletenni egy visszacsatolást.
- A matching algoritmust

**Begyűjtendő információ**
- Két megfigyelés típus van, az önkéntesek és a diákok, különböző információt kell tőlük bekérni.

*Önkéntesek:*
- Alapvető adatok: Név, Email, Nem, Egyetem, ahol tanulsz, Rövid bemutatkozás

Matchingeléshez szükséges információk:

- Milyen tárgyból van tudása? (általános: matek, töri, közgáz, vezszerv - attól függ kire lő a program)
- Milyen témákból van tudása? (specifikusabb: mateknál: algebra pl)
- Milyen mélyen tudsz segíteni (0: egy-két feladatban - 5: expert vagyok/amúgy is tanítok diákokat)
- Milyen rendszerességgel tud segíteni? (Csak egyszeri, vagy hetente)
- Mikor ér rá? (napok és azn belül is napszakok megadása lenne talán a célszerű: szerda délután pl, szerda koraeste)

*Diákok:*
- Alapvető információk: Név, Email, Nem, Egyetem/középiskola, ahol tanul, Rövid bemutatkozás

Matchinegléshez szükséges információk: 
- Milyen tárgyból kell segítség?
- Milyen részterületről kell segítség?
- Milyen szinten tudod az anyagot? (0-5, ahol a 0 az egyáltalán nem értem, az öt pedig a nem jön ki egy feladat kb)
- Vizsgára készüls vagy érettségire? Meddig van szükséged segítségre?
- Milyen gyakran kellene a segítség?
- Mikor ér rá? (napok és azn belül is napszakok megadása lenne talán a célszerű: szerda délután pl, szerda koraeste)

**Tervezett lépések**
    
1) Kitölteni a spreadet
    
2) Algoritmus keres egy párt, aki pár tolodban teljesen (mikor ér rá) pár dologban elég, ha csak részben matchel (pontos topicok)
    
3) Ha talál párt, akkor küld emailt mind a két félnek a másik alap infóiról.


## RAJK50 kutatás

**Rövid leírás**: A Rajk50 kutatást megtámogatva létrehozni egy interaktív felületet, ahol vizualizáljuk az eredményeket. 

**begyűjtött adat struktúrája**
- SPSS-ben van meg az adattábla, ami tartalmazza a felvételi adatokat + voszkos ki-kicsodát, ezt könnyen át lehet emelni egy Dataframebe
- Ez egy datab tábla, kb 2500 mefigyeléssel, ami nem nagy
 
**frissítési rendszeresség és a frissítés menete**
- Voszkos kikicsoda frissül pár évente
- Felvételi adatbázis: Nincsen tervbe, hogy minden évben frissüljön, de talán rá lehet torzítani a kollégiumot
- frissülés esetén több megfigyelésünk lesz
 
**milyen értékes információt lehet a begyűjtött anyagból egy dashboardon kommunikálni**
- User interakció a vizualizációnál lehet, a user választhatja ki, mit akar látni a táblákon
- Ahogy a kutatásban, itt is szét lehet bontani a megfigyeléseket - felvett nem felvettre, voszkos lett - kilépettre; illetve minden időszak szerint. 
    - Térképen a származást, mostani lakhelyet
    - Diagrammon az, hogy ki milyen végzettségű szülők gyereke
    - Diagrammon a nemek eloszlása
    - Pozíciók eloszlása nemek, karon, stb szerint. 
    - (minden egyéb, amit  a Rajk50-esek vizsgáltak)
